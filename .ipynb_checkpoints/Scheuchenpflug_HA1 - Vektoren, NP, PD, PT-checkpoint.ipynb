```python
import pandas as pd
import numpy as np
import torch as pt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import os
import sys
```

```python
ex_vector = [4, 6, 9, -1]
# random example vector
```

#### Vector representation in different frameworks

```python
# vector in Numpy
np_vector = np.array(ex_vector)
print(np_vector, "\nbeing ", type(np_vector))
print("dimension: ", np_vector.size)
```

```python
# vector in Pandas
pd_vector = pd.Series(np_vector)
print(pd_vector, "\nbeing", type(pd_vector))
print("dimension: ", pd_vector.size)
```

```python
# vector in PyTorch
pt_vector = pt.tensor(pd_vector)
print(pt_vector, "\nbeing", type(pt_vector))
print(pt_vector.size())
```

#### Age vectors from kaggle dataset "Titanic"

```python
titanic_df = pd.read_csv(".\\kaggle_titanic\\full.csv")
titanic_df.head()
```

```python
ages = pd.Series(titanic_df["Age_wiki"])
# replace NaN values with 0, otherwise mean() and manual approach behave differently
ages = ages.fillna(0.00)
```

#### Novellenschatz

```python
data = pd.DataFrame(columns=["title", "text"])
for file in os.listdir(r"./novellenschatz/full/"):
    with open(rf"./novellenschatz/full/{file}", "r", encoding="utf8") as f:
        data = data.append({'title':file, 'text':f.read()}, ignore_index=True)
```

```python
data.head()
```

```python
cv = CountVectorizer(input="content")
X = cv.fit_transform(data["text"])
```

```python
np.set_printoptions(threshold=sys.maxsize)
cv.get_feature_names_out()
```

```python
# extra steps, don't know why
full_array = X.toarray()
X_df = pd.DataFrame(full_array)
term_pos = cv.vocabulary_.get("das")
X_df[term_pos].sum()
```

```python
#make DF for tf/tfidf values
out = pd.DataFrame(columns=["sum", "das", "tf", "tfidf"])
out["sum"] = X_df.sum(axis=1)
out["das"] = X_df[term_pos]
out["tf"] = out["das"]/out["sum"]
docf = sum(X_df[term_pos] > 0)
out["tfidf"] =  out["tf"]*np.log(X_df.shape[0]/docf+1)
out.head()
```

```python
tfidfv = TfidfVectorizer(use_idf=True)
Y = tfidfv.fit_transform(data["text"])
```
